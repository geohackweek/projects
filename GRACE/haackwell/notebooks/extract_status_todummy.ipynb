Extract and clean the Status field

```python
# clean up statements
dummy = wells.STATUS.str.lower() # lowercase phrases
dummy = dummy.str.replace('\r\n','') # remove hard newlines
dummy = dummy.str.replace('\n','') # remove soft newlines
dummy = dummy.str.replace('|',' ') # remove the |
dummy = dummy.str.replace(':','. ') # remove the :
dummy = dummy.str.replace(',','. ') # remove the ,
dummy = dummy.str.replace(' - ',' ') # remove the floating hyphen
dummy = dummy.str.replace('- ','') # remove the floating hyphen
dummy = dummy.str.replace('-','') # remove the floating hyphen
dummy = dummy.str.replace('(','. ').str.replace("."," ") # remove the ()
dummy = dummy.str.replace('\s+',' ') # remove surplus spaces
```

Construct the dataframe for each of the binary vectors

```python
def dicttomatrix(dictionary):
    phrasedf = pd.DataFrame()

    for eachtuple in dictionary:
        phrasedf[eachtuple[0]]=pd.Series([int(i in eachtuple[1]) for i in range(0, eachtuple[1][-1]+1)])

    # binary map for the phrases
    phrasedf = phrasedf.fillna(0)
    return phrasedf
```

```python
# mech_ind
outdf1['adminfail'] = outdf1.loc[:, admin_0].fillna(0).apply(lambda x: 1 if sum(x)>0 else 0, axis=1)
outdf1['mechfail'] = outdf1.loc[:, mech_1].fillna(0).apply(lambda x: 1 if sum(x)>0 else 0, axis=1)
outdf1['envfail'] = outdf1.loc[:, env_2].fillna(0).apply(lambda x: 1 if sum(x)>0 else 0, axis=1)
outdf1['pollutionfail'] = outdf1.loc[:, pollution_3].fillna(0).apply(lambda x: 1 if sum(x)>0 else 0, axis=1)

# All failure markers to csv

wellfails = wells.merge(pd.DataFrame(outdf1.loc[:, ['adminfail', 'mechfail', 'envfail', 'pollutionfail']]), how='left', left_index=True, right_index=True)
wellfails.to_csv('/Users/ymseah/Google Drive/GeoHackWeek/haackwell/dat/wellfails.csv', doublequote=False, index=False)
```
